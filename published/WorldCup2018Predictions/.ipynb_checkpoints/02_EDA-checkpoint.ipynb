Mean value for the goal difference is greater than 0.7, which may present a problem later on when training the model as it could lead to a bias of team1 being greather than team2.

To get rid oif this, randomize the order in which teams are listed for any one match. 

```python
random.seed(555)
switch = []
for x in range(len(df)):
    if random.random() > 0.5:
        switch.append(1)
    else:
        switch.append(0)
df['switch'] = switch
```

```python
for ii in range(len(df)):
    if (df['switch'][ii]==1):
        orig_team1 = df.loc[ii,'team1_edit']
        orig_team2 = df.loc[ii,'team2_edit']
        orig_team1_score = df.loc[ii,'team1_score']
        orig_team2_score = df.loc[ii,'team2_score']
        orig_score_only = df.loc[ii,'score_only']
        df.loc[ii,'team1_edit'] = orig_team2
        df.loc[ii,'team2_edit'] = orig_team1
        df.loc[ii,'team1_score'] = orig_team2_score
        df.loc[ii,'team2_score'] = orig_team1_score
        df.loc[ii,'score_only'] = str(df.loc[ii,'team1_score'])+'-'+str(df.loc[ii,'team2_score'])
df['goal_diff'] = df['team1_score'] - df['team2_score']
df.drop(columns='switch',inplace=True)
```

## Performance table
Need to look at each team's performance over its history, so I will need a team-centric dataset.

Building this dataset is simple: take each observation in matches - which has the form "team1 vs team2" - and produce two separate observations of the form "team1 played against team2" and "team2 played against team1" respectively.

```python
def gameRound(gr):
    if game_round == gr:
        return 1
    else:
        return 0
def isWin(Team1,Team2):
    if Team1 > Team2:
        return 1
    else:
        return 0
def isLoss(Team1,Team2):
    if Team1 < Team2:
        return 1
    else:
        return 0
def isDraw(Team1,Team2):
    if Team1 == Team2:
        return 1
    else:
        return 0    

perf_table = []
for ii in range(len(df)):
    # fixed items
    edition = df.loc[ii,'edition']
    game_round = df.loc[ii,'round']
    year = df.loc[ii,'year']
    competition = df.loc[ii,'competition']
    pseudo_date = df.loc[ii,'pseudo_date']
    match_id = df.loc[ii,'match_id']
    #variables
    team1_iso = df.loc[ii,'team1_iso']
    team2_iso = df.loc[ii,'team2_iso']
    team1_score = int(df.loc[ii,'team1_score'])
    team2_score = int(df.loc[ii,'team2_score'])
    score_only = df.loc[ii,'score_only']
    # row 1
    r1 = [match_id, edition, year, competition,pseudo_date,
          gameRound('PRELIMINARY'), gameRound('GROUP_STAGE'), gameRound('1/8_FINAL'),
          gameRound('1/2_FINAL'), gameRound('1/4_FINAL'), gameRound('FINAL'),
          team1_iso, team2_iso, team1_score, team2_score, team1_score - team2_score,
          isWin(team1_score,team2_score),isLoss(team1_score,team2_score),isDraw(team1_score,team2_score) ]
    # row 2
    r2 = [match_id, edition, year,competition,pseudo_date,
          gameRound('PRELIMINARY'),gameRound('GROUP_STAGE'),gameRound('1/8_FINAL'),
          gameRound('1/2_FINAL'),gameRound('1/4_FINAL'), gameRound('FINAL'),
          team2_iso, team1_iso, team2_score, team1_score, team2_score - team1_score,
          isWin(team2_score,team1_score),isLoss(team2_score,team1_score),isDraw(team2_score,team1_score) ]
    # group together
    headers = ['match_id','edition','year','competition','pseudo_date',
               'PRELIMINARY','GROUP_STAGE','1/8_FINAL',
               '1/2_FINAL', '1/4_FINAL','FINAL',
               'team','opponent', 'scoreTeam','scoreOpponent','goaldiff',
               'WIN','LOSS','DRAW']
    perf_table.append(dict(zip(headers,r1)))
    perf_table.append(dict(zip(headers,r2)))    
df_performance_table = pd.DataFrame(perf_table)
df_performance_table['total_goals'] = df_performance_table['scoreTeam'] + df_performance_table['scoreOpponent']
df_performance_table.drop_duplicates(inplace=True)
len(df_performance_table)
```

# Remove outliers


Since our aim is to predict the goal differential (win margin) between the two teams in a match, we'd like to get rid of outliers - values which are far away at the end of the spectrum of possible values for this variable. The reason is that outliers can drastically change the results of the data analysis and statistical modeling. Outliers increase the error variance, reduce the power of statistical tests, and ultimately they can bias or influence estimates.
#### As a rule of thumb, any value which is out of range of the 5th and 95th percentile may be considered an outlier.
In my case this, is 7.


```python
np.percentile(df_performance_table['total_goals'],95)
```

```python
df = df[abs(df['total_goals']) < 7]
df.drop_duplicates(inplace=True)
print(len(df))

df_performance_table = df_performance_table[abs(df_performance_table['total_goals']) < 7]
df_performance_table.drop_duplicates(inplace=True)
print(len(df_performance_table))
```

## Strength of opposition
We may also want to take into account the fact that teams play most of their matches against opponents from the same FIFA Confederation - for example, European teams play mostly against other UEFA members, while African teams face other CAF members for the most part. Only during final tournaments like the Olympics, the World Cup and the Confederations Cup will teams play official (non-friendly) matches against non-confederation opponents.

Since not all conferences are the same general strength, we can adjust our teamperf dataset to also include information about the conference the opponent belongs to. We can assign adjustment coefficients to each conference, in a similar way to how FIFA's World ranking algorithm accounts for regional strength.

```python
# confederations and adjustment coefficients for them
conf = ["UEFA","CONMEBOL","CONCACAF","AFC","CAF","OFC"]
adj = [0.99,1,0.85,0.85,0.85,0.85]
df_conf = pd.DataFrame.from_records(list(zip(conf,adj)),columns = ['Confederation','Adjustment'])
```

```python
# add coefficient to opponent
df_conf_map = df_teams[df_teams.columns.intersection(['confederation','iso'])]
df_conf_map.drop_duplicates(inplace=True)

df_performance_table = pd.merge(df_performance_table,df_conf_map,how='left',left_on='team',right_on='iso')
df_performance_table.drop(columns='iso', inplace=True)

len(df_performance_table)
```

```python
df_performance_table = pd.merge(df_performance_table,df_conf,how='left',left_on='confederation',right_on='Confederation')
df_performance_table.drop(columns=['confederation'],inplace=True)
```

# Feature Engineering
Now, let's calculate some lag features for each team which is about to play a game.

We'll look at the previous N games a team has played, up to the game in question, and we'll calculate the percentage of wins, draws, losses, as well as the goal differential, per game, for those past N games.

For example, taking N=10:

last10games_w_per = (number of wins in the past 10 games) / 10
last10games_d_per = (number of draws in the past 10 games) / 10
last10games_l_per = (number of losses in the past 10 games) / 10
last10games_gd_per = (goals scored - goals conceeded in the past 10 games) / 10
We'll use three different values for N (10, 30 and 50) to capture short-, medium-, and long-term form.

We'll calculate those values for every team and every game in our dataset.

To model the strength of opposition faced, we'll use the same technique with respect to the opponentConfederationCoefficient values we introduced earlier.

```python
df_performance_table.sort_values('pseudo_date',ascending=True, inplace=True)
```

```python
def lastNgames(COL,WINDOW):
    xx = df_performance_table.groupby(['team'])[COL].rolling(WINDOW).sum().reset_index().set_index('level_1').reset_index().sort_values('level_1')[COL] / WINDOW
    return list(np.nan_to_num(xx))
df_performance_table['last10_w'] = lastNgames('WIN',10)
df_performance_table['last20_w'] = lastNgames('WIN',20)
df_performance_table['last40_w'] = lastNgames('WIN',40)

df_performance_table['last10_l'] = lastNgames('LOSS',10)
df_performance_table['last20_l'] = lastNgames('LOSS',20)
df_performance_table['last40_l'] = lastNgames('LOSS',40)

df_performance_table['last10_d'] = lastNgames('DRAW',10)
df_performance_table['last20_d'] = lastNgames('DRAW',20)
df_performance_table['last40_d'] = lastNgames('DRAW',40)

df_performance_table['last10_gd'] = lastNgames('goaldiff',10)
df_performance_table['last20_gd'] = lastNgames('goaldiff',20)
df_performance_table['last40_gd'] = lastNgames('goaldiff',40)

df_performance_table['last10_opp_per'] = lastNgames('Adjustment',10)
df_performance_table['last20_opp_per'] = lastNgames('Adjustment',20)
df_performance_table['last40_opp_per'] = lastNgames('Adjustment',40)
```

Now that we have built a series of team-specific features, we need to fold them back into match-specific features.

We will then have a set of features for both teams about to face each other.


```python
labels =['match_id','team','opponent',
         'last10_w', 'last20_w', 'last40_w', 
         'last10_l', 'last20_l', 'last40_l',
         'last10_d', 'last20_d', 'last40_d', 
         'last10_gd', 'last20_gd', 'last40_gd', 
         'last10_opp_per', 'last20_opp_per', 'last40_opp_per']
df_performance_table_map = df_performance_table[df_performance_table.columns.intersection(labels)]
df_match_team_perf = pd.merge(df,df_performance_table_map, how='inner',
                              left_on=['match_id','team1_iso', 'team2_iso'],
                              right_on=['match_id','team','opponent'])
```

```python
# how correlated are these features?
labels = ['goal_diff','last10_w', 'last20_w', 'last40_w', 'last10_l', 'last20_l',
       'last40_l', 'last10_d', 'last20_d', 'last40_d', 'last10_gd',
       'last20_gd', 'last40_gd', 'last10_opp_per', 'last20_opp_per',
       'last40_opp_per']
df_corr = df_match_team_perf[df_match_team_perf.columns.intersection(labels)]
```

```python

import seaborn as sns
%matplotlib inline

# calculate the correlation matrix
corr = df_corr.corr()

# plot the heatmap
sns.heatmap(corr, 
        xticklabels=corr.columns,
        yticklabels=corr.columns)
```

```python
cmap = cmap=sns.diverging_palette(5, 250, as_cmap=True)

def magnify():
    return [dict(selector="th",
                 props=[("font-size", "7pt")]),
            dict(selector="td",
                 props=[('padding', "0em 0em")]),
            dict(selector="th:hover",
                 props=[("font-size", "12pt")]),
            dict(selector="tr:hover td:hover",
                 props=[('max-width', '200px'),
                        ('font-size', '12pt')])
]

corr.style.background_gradient(cmap, axis=1)\
    .set_properties(**{'max-width': '80px', 'font-size': '10pt'})\
    .set_caption("Hover to magify")\
    .set_precision(2)\
    .set_table_styles(magnify())
```

Not a surprise here, goal difference is strongly correlated to Win Values and strongly X-correlated to Loss Values.

Also, we have strong positive correlations between lag features measuring the same metric on different lag windows. For example, the last10_win, last20_win and last50_win metrics are correlated. Also unsurprisingly, the correlation between longer windows is weaker than for closer windows. But it does seem to suggest that good teams generally keep being good over time, and bad teams keep being bad.
