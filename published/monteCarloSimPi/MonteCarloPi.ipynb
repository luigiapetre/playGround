# Approximate $\pi$ numerically using Monte Carlo Simulations
## Background
Consider the following unit circle $r=1$ inscribed within a square
![fig_1](http://mathfaculty.fullerton.edu/mathews/n2003/montecarlopi/MonteCarloPiMod/Images/MonteCarloPiMod_gr_5.gif)

The area of the cirlce is $$\pi r^2 = \pi \times 1^2 = \pi,$$ and the area of the square is $$(2r)^2 = 2^2 = 4.$$

The ratio of the two areas is thus:
$$\rho = \frac{\text{Area of Circle}}{\text{Area of Square}} = \frac{\pi r^2}{(2r)^2} = \frac{\pi}{4}$$

So, if we want to approximate $\pi$, this can be found as
**$$ \pi = 4\times\rho$$**

### So how can we find $\rho$? 
Consider the following:
![fig_2](http://mathfaculty.fullerton.edu/mathews/n2003/montecarlopi/MonteCarloPiMod/Images/MonteCarloPiMod_gr_11.gif)

Using Monte Carlo simulations, for say $n$ experiments. If I had to randomly select pairs $(x_i,y_i)$ for $i = 1,...,n$ in the unit square, then the ratio can be deteremined as 
$$\rho = \frac{m}{n} $$ where $m$ is number of points within the unit circle and $n$ is the total number of experiments. 

To identify these points $m$, we make use of elementary geometry in the form of pythagorean theory such that, the pair $$(x_i,y_i) \text{ fall within $m$} \iff x_i^2+y_i^2\leq1^2 \equiv 1 $$

However, every time a Monte Carlo simulation is made using the same sample size $n$ it will come up with a slightly different value. No worries. The values converge very slowly of the order $o(n^{-1/2})$ as a direct consequence of the [Central Limit Theorem](https://en.wikipedia.org/wiki/Central_limit_theorem).

For simplicity, I will only consider the first quadrant of the circle, shown below.
![fig_3](http://mathfaculty.fullerton.edu/mathews/n2003/montecarlopi/MonteCarloPiMod/Images/MonteCarloPiMod_gr_25.gif)

## Solution

```python
import random
import time
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import unicodedata
from math import pi
```

```python
def withinCircle(x,y):
    """
    test to see if the pair falls within unit cirlce
    """
    if (x**2 + y**2 ) <= 1:
        return 1
    else:
        return 0
def monteCarloPI(N):
    X = []
    Y = []
    R = []

    t0 = time.time()
    for n in range(N):
        x = random.uniform(0, 1)
        y = random.uniform(0, 1)
        r = withinCircle(x,y)
        X.append(x)
        Y.append(y)
        R.append(r)
    t1 = time.time() - t0
    PI = 4*float(sum(R))/(N)
    return {'X': X,
           'Y': Y,
           'R': R,
           'timetaken': t1,
           'PI' : PI}
```

```python
N = 10000 #number of experiments
```

```python
MC = monteCarloPI(N)
X = MC['X']
Y = MC['Y']
R = MC['R']
t1 = MC['timetaken']
PI = MC['PI']

print('Approximation of {}:'.format(unicodedata.lookup("GREEK SMALL LETTER PI")))
print('     After {} iterations lasting {:.5f} secs: {:.10f}'.format(N,t1,PI))
print('     Analytical value:  {:.10f}'.format(pi))
print('     Inaccuracy: {:.3}%'.format(((pi-PI)/pi) * 100))
```

```python
df = pd.DataFrame({'X':X,'Y':Y,'R':R})

fig, ax = plt.subplots(figsize=(6, 6))
ax.plot(df['X'][df['R']==0],df['Y'][df['R']==0], '.b', label='Outside')
ax.plot(df['X'][df['R']==1],df['Y'][df['R']==1], '.r', label='Inside')
leg = ax.legend(loc='lower center');
ax.set_xlim([-0.1,1.1])
ax.set_ylim([-0.2,1.1])
ax.set_aspect('equal')
plt.show()
```

# Can we prove convergence?

```python
NN = 7
MC = []
for enum, n in enumerate(np.geomspace(1,10**NN, num=NN+1, dtype=int)):
    print('---{} - Started MC {}'.format(enum,n))
    MC.append([n, monteCarloPI(n)])
```

```python
M = [x[0] for x in MC]
T = [ x[1]['timetaken']   for x in MC]
PI = [ x[1]['PI']   for x in MC]
ERROR = [abs((float(p)-pi))/pi for p in PI]
```

```python
fig, (ax1,ax2,ax3) = plt.subplots(3, sharex=True,figsize=(12, 10))
ax1.semilogx(M,[t/60 for t in T])
ax1.set_yscale('log')
ax1.set_title("Time taken in Minutes")

ax2.semilogx(M,ERROR)
ax2.set_yscale('log')
ax2.set_title('Error between Analytical and Numerical {}:'.format(unicodedata.lookup("GREEK SMALL LETTER PI")) )

ax3.semilogx(M,PI,'c',label = 'Numerical')
ax3.semilogx(M, len(M)*[pi],'m', label = 'Analytical')
ax3.set_yscale('log')
ax3.legend()
ax3.set_title('Convergence of Numerical to Analytical Solution')

plt.show()
```
