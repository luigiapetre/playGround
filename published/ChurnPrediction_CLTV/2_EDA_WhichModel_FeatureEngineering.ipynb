# 2_EDA_WhichModel_FeatureEngineering
In this notebook, I perform an EDA to try and understand which model would be most suited for the provided data. The problem at hand is to **increase casino net revenue**. i.e. we would like to maximise the NGR per player, where NGR is defined as $$NGR = Bets – Wins – Bonus Cost – Tax$$

I consider different visualisations, employ feature engineering and gauge different perspectives of the data to come up with a strategy to try and solve for this.

### OK, so based on the above, I shall propose the following strategy to increase NGR
Given a customer who transacted with us for the 1st time.
1. **10 day mark:** After 10 days since the player's first transaction, run the customer through a churn prediction model which will predict if a churn happened in the past 10 days. I chose 10 since (i) it is longer than a week, thus we can gauge a player's weekly pattern and at least 1 weekend, and (ii) it encapsulates **25%** of all our customers, or **37%** when exclude Day 0 players. **THIS IS A WORRY**. 

    From the predicted churn outcome,
    - **if the customer did not churn or will not churn within 10 days**:
        - either do not apply any change to the current bonuses, promo, etc, or
        - implement bonus savings on these players. As a first pass of the model, I would advise against this for now since I would like to focus on increasing player retention at the moment.
    - **if the customer is going to churn or has already churned**: 
        - pass through a NGR prediction (early value classification model) to identify leads. Promos and bonuses are then sent to the high value clients to try and retain or reactivate them.
1. **30 day mark:** 
    - Pass the players through a churn prediction model tuned for day 29, 30, 31 days.
    - Apply the previous logic to identify promo and discount distribution

````python
def addChurn(colName,minDay,maxDay):
    churn = []
    for x in df_players['DIFF_trans_max_min']:
        if (x >= minDay) & (x <=maxDay):
            churn.append(1)
        else:
            churn.append(0)
    df_players[colName] = churn
addChurn('churn_10',1,10)
addChurn('churn_30',29,31)
````

````python
def getDominantDayOfWeekByCol(col, dtt, dt, selectFunc):
    '''Return the DayOfWeek based on the the selectFunc(min-mix) for col '''
#     col = 'count_deposit'
#     X = dtt[['DayOfWeek',col]]
#     selectFunc = 'min'
    X = dtt[['DayOfWeek',col]]

    A = []
    # pick DayOfWeek mode
    try: 
        st.mode(X['DayOfWeek'])
        A.append(st.mode(X['DayOfWeek']))
    # if multiple days have the same mode, pick the one with the largest aggregated deposit counts    
    except:
        if dtt.shape[0] == 0:
            A.append( [d.weekday() for d in dt['date_transaction_id'].head(1)])
        else:
            # dataframe within the mode Day Of Week
            B = pd.DataFrame(sorted( Counter(X['DayOfWeek']).items() ))
            B.columns = ['DayOfWeek','Freq']

            # dataframe with the count_deposit for all days of week, aggregated for Days
            C = X[['DayOfWeek',col]] #df with necessary fields
            C = C.groupby(['DayOfWeek']).agg('sum').reset_index() # aggregation
            if selectFunc == 'min':
                C = C[C[col] == min(C[col])] # select maximum frequency deposit count
            elif selectFunc == 'max':
                C = C[C[col] == max(C[col])] # select maximum frequency deposit count 

            # inner join to get only the relevent DayOfWeek
            D = pd.merge(B,C, how='inner', on='DayOfWeek')

            A.append(D['DayOfWeek'][0])

    return A 
````

````python
def getCustomerMetric(cid,dayRange):
    ''' Return Customer Metrics for a customerid (cid) for a predefined day range (dayRange)'''
    # cid = 200
    # dayRange = 10
    # whilst interating, print progress for the nth cid
    if cid%200 == 0:
        print('DayRange {} : {} [{:00%}]'.format(dayRange, cid, cid/max(df['customer_id'])))

    # select the customer
    dt = df[df['customer_id'] == cid]
    # get transaction history for the past dayRange days
    dtt = dt[ (dt['date_transaction_id'] >= min(dt['date_first_deposit_id'])) 
             & (dt['date_transaction_id'] <= min(dt['date_first_deposit_id']) + timedelta(days = dayRange)) ]
    if dtt.shape[0] == 0:
        dt_res = pd.DataFrame({'customer_id' : [164]})
    else:
        # get player timing differences MIN MAX
        dtt_minmax = dtt[['customer_id','date_registration_id', 'date_first_deposit_id','date_transaction_id']].groupby(['customer_id']).agg([('Min','min'), ('Max','max')]).reset_index()
        dtt_minmax.columns = [' '.join(col).strip() for col in dtt_minmax.columns.values]
        dtt_minmax.columns = dtt_minmax.columns.get_level_values(0)
        dtt_minmax['DIFF_{}_deposit_reg'.format(dayRange)] = [d.days for d in dtt_minmax['date_first_deposit_id Min'] - dtt_minmax['date_registration_id Min']]
        dtt_minmax['DIFF_{}_trans_deposit'.format(dayRange)] = [d.days for d in dtt_minmax['date_transaction_id Min'] - dtt_minmax['date_first_deposit_id Min']]
        dtt_minmax['DIFF_{}_trans_max_min'.format(dayRange)] = [d.days for d in dtt_minmax['date_transaction_id Max'] - dtt_minmax['date_transaction_id Min']]
        dtt_minmax.drop(['date_registration_id Min','date_registration_id Max','date_first_deposit_id Min',
                         'date_first_deposit_id Max','date_transaction_id Min','date_transaction_id Max'], axis=1, inplace=True)    

        # aggregate SUM variables
        dtt_sum = dtt.groupby(['customer_id']).agg('sum').reset_index()
        dtt_sum.drop(['acquisition_source','Country_Lat','Country_Long'],axis=1, inplace=True)
        dtt_sum.columns = ['TOTAL_{}_'.format(dayRange) + str(col) for col in dtt_sum.columns.tolist() ]
        dtt_sum.rename(columns={'TOTAL_{}_'.format(dayRange) + 'customer_id':'customer_id'}, inplace=True)

        ### BEHAVIOUR METRICS WITHIN THE DAY RANGE ###
        #### Absolute ####
        # Count of days using Deposits, FreeSpin, RealBets, Bonus
        dtt_sum['COUNT_{}_{}'.format(dayRange, 'numDeposits')] = dtt[dtt['count_deposit']>0].shape[0]
        dtt_sum['COUNT_{}_{}'.format(dayRange, 'numFreeSpin')] = dtt[dtt['sum_free_spin']>0].shape[0]
        dtt_sum['COUNT_{}_{}'.format(dayRange, 'numRealBets')] = dtt[dtt['sum_bet_real']>0].shape[0]
        dtt_sum['COUNT_{}_{}'.format(dayRange, 'numBonusBets')] = dtt[dtt['sum_bet_bonus']>0].shape[0]

        #### iGaming KPIs ####
        # NGR_Deposit_ratio: 
        #      Low values might indicate a high number of winners or abusers.
        #      High values are showing players spending more money than usual.
        dtt_sum['RATIO_{}_{}'.format(dayRange, 'NGR_Deposit')] = dtt_sum['TOTAL_{}_'.format(dayRange) + 'sum_ngr'] / dtt_sum['TOTAL_{}_'.format(dayRange) + 'sum_deposit']
        # Bets_Deposit_ratio: 
        #      How many times players circulate each money they have deposited.
        #      Higher figures mean that it takes more time to lose whilst lower results indicate that players are losing
        #      at a faster pace. 
        dtt_sum['RATIO_{}_{}'.format(dayRange, 'Bets_Deposit')] = dtt_sum['TOTAL_{}_'.format(dayRange) + 'sum_bet_real'] / dtt_sum['TOTAL_{}_'.format(dayRange) + 'sum_deposit']
        dtt_
        

        #### Modified RFM Analysis in the form of RFMI ####
        # R = Recency : How recently did the customer interact?
        X = dtt[['customer_id','date_transaction_id']].groupby(['customer_id']).agg('max').reset_index()
        dtt_sum['RFM_{}_'.format(dayRange) +'R'] = [d.days for d in pd.datetime.now() - X['date_transaction_id'] ]
        # F = Frequency  : How often did they purchase within their lifetime?
        # M = Monetary : How much do they spend?
        #  inhereted from 'TOTAL_...' and 'COUNT_...' columns
        # I = Interactions: To assess the count of engagement with us as well the deterring effect of marketing engagements that are too frequent or too concentrated in time
        dtt_sum['RFM_{}_'.format(dayRange) +'I'] = dtt.shape[0]
        #  RE the deterring effect of marketing engagement: this is inhereted from the 'TOTAL_...' and 'COUNT_...' 
        #     columns relating to free spin bonuses

        #### When is the customer most likely to interact with us ####
        # pick the player mode day to play
        # if multiple days have the same mode, pick the one with the largest aggregated deposit counts    
        dtt['DayOfWeek'] = [d.weekday() for d in dtt['date_transaction_id'] ]
        A = []
        # pick DayOfWeek mode
        try: 
            A.append(st.mode(dtt['DayOfWeek']))
        # if multiple days have the same mode, pick the one with the largest aggregated deposit counts    
        except:    
            if dtt.shape[0] == 0:
                A.append( [d.weekday() for d in dt['date_transaction_id'].head(1)])
            else:
                # dataframe within the mode Day Of Week
                B = pd.DataFrame(sorted( Counter(dtt['DayOfWeek']).items() ))
                B.columns = ['DayOfWeek','Freq']

                # dataframe with the count_deposit for all days of week, aggregated for Days
                C = dtt[['DayOfWeek','count_deposit']] #df with necessary fields
                C = C.groupby(['DayOfWeek']).agg('sum').reset_index() # aggregation
                C = C[C['count_deposit'] == max(C['count_deposit'])] # select maximum frequency deposit count

                # inner join to get only the relevent DayOfWeek
                D = pd.merge(B,C, how='left', on='DayOfWeek')

                A.append(D['DayOfWeek'][0])
        dtt_sum['BEH_{}_DOW_Interaction'.format(dayRange)] = A

        #### Other likely Day Of Week features ####
        #most likely day to have 'max' of 'count_deposit'
        dtt_sum['BEH_{}_DOW_count_deposit'.format(dayRange)] = getDominantDayOfWeekByCol('count_deposit', dtt, dt, 'max') 
        # other BEH features:
        dtt_sum['BEH_{}_DOW_count_deposit'.format(dayRange)] = getDominantDayOfWeekByCol('count_deposit', dtt, dt, 'max') 
        dtt_sum['BEH_{}_DOW_bet_real'.format(dayRange)] = getDominantDayOfWeekByCol('sum_bet_real', dtt, dt, 'max') 
        dtt_sum['BEH_{}_DOW_bet_bonus'.format(dayRange)] = getDominantDayOfWeekByCol('sum_bet_bonus', dtt, dt, 'max') 
        dtt_sum['BEH_{}_DOW_free_spin'.format(dayRange)] = getDominantDayOfWeekByCol('sum_free_spin', dtt, dt, 'max') 
        dtt_sum['BEH_{}_DOW_win_real'.format(dayRange)] = getDominantDayOfWeekByCol('sum_win_real', dtt, dt, 'min') 
        dtt_sum['BEH_{}_DOW_win_bonus'.format(dayRange)] = getDominantDayOfWeekByCol('sum_win_bonus', dtt, dt, 'min') 

        #### Averages ##### 
        # What is the player's Avg deposit
        dtt_sum['AVG_{}_deposit'.format(dayRange)] = dtt_sum['TOTAL_{}_sum_deposit'.format(dayRange)] /  dtt_sum['TOTAL_{}_count_deposit'.format(dayRange)] 
        # Avg based on the total transactions with us
        dtt_avg = dtt.groupby(['customer_id']).agg('mean').reset_index()
        dtt_avg.drop(['acquisition_source','Country_Lat','Country_Long','DayOfWeek'],axis=1, inplace=True)
        dtt_avg.columns = ['AVG_{}_{}'.format(dayRange,str(col)) for col in dtt_avg.columns.tolist() ]
        dtt_avg.rename(columns={'AVG_{}_{}'.format(dayRange,'customer_id'):'customer_id'}, inplace=True)

        # Merge DataFrames with results    
        dt_res = pd.merge(dtt_sum, dtt_avg, how='left', on='customer_id')
        dt_res = pd.merge(dt_res, dtt_minmax, how='left', on='customer_id')

    return dt_res
````

````python
def getNGR_Split(col, W):
#     col = 'TOTAL_10_sum_ngr'
#     W = df_10[['customer_id',col]]
    print('-------> {} : {}'.format(col, W.shape))
    #  for -ve NGR, segregate to class 0
    Y = W[W[col]<0]
    Y['CLTV'] = [0]* Y.shape[0]

    # Split the remaining +ve NGRs based on 3 quantiles [1, 2, 3], with 
    #     1 being the low value NGR,
    #     3 being the high value NGR
    X = W[W[col]>=0]
    X['CLTV'] = pd.qcut( X[col], q=3, labels=[1,2,3] )
    CLTV_intervals = set(pd.qcut( X[col], q=3 ))
    X['CLTV'] = [ int(cltv) for cltv in X['CLTV'] ]
    
    # concat both df
    df_cltv = pd.concat([X,Y], sort=False)
    df_cltv.sort_values(['customer_id'], inplace=True)

    print('----------> -ve NGR {} : +ve NGR {} : df_cltv {} '.format(Y.shape, X.shape, df_cltv.shape))
    if (W.shape[0] == X.shape[0] + Y.shape[0]):
        print('----------> Shape in agreement for {} : {} == {}'.format(col, df_cltv.shape[0], X.shape[0] + Y.shape[0]))
    else:
         raise ValueError('Shape not in agreement')
    
    return df_cltv
````

````python
def getCustomerMetricDF(numDays):
#     numDays = 10
    t0 = time.time()
    X = []
    for x in df_players['customer_id'] :
        X.append(getCustomerMetric(x,numDays))
        if (x%600==0):
            print("-------> Time for {} customer_id's : {:.2f} min".format( x, (time.time() - t0) / 60 ) )
    print("-------> Time for {} customer_id's : {:.2f} min".format( len(df_players['customer_id']), (time.time() - t0) / 60 ) )
    X = pd.concat(X, axis=0,  ignore_index=True, sort=False)
    print("-------> Time after Concatenation of cid : {:.2f} min".format( (time.time() - t0) / 60 ) )
    print('Shape after metric creation : {}'.format(X.shape))

    # inpute nan values with median  
    X.fillna(X.median(), inplace=True)
    print('Shape after inpute : {}'.format(X.shape))

    # calculate NGR split. 
    # This is based on the work highlighted further on in the NGR section. I am pre-empting this part since it
    # makes more sense to contain the split within the function
    print("Started NGR split calculation for df_{}".format( numDays) )
    colNGR = ['TOTAL_'+str(numDays)+'_sum_ngr', 'AVG_'+str(numDays)+'_sum_ngr']
    Y = []
    for y in colNGR:
        Y.append(getNGR_Split(y,X[['customer_id',y]]))
    Y = pd.concat(Y, axis=1, ignore_index=True, sort=False)
    Y.drop([3, 4], axis=1, inplace=True)
    Y.columns = ['customer_id','GROSS_'+str(numDays)+'_sum_ngr',
                 'NGR_SPLIT_TOTAL_'+str(numDays)+'_sum_ngr', 'NGR_SPLIT_AVG_'+str(numDays)+'_sum_ngr']
    # merge X & Y
    print('Shape X {}'.format(X.shape))
    print('Shape Y {}'.format(Y.shape))
    Y.rename(columns={'customer_id':'customer_id_Y'}, inplace=True)
    Y.drop(['customer_id_Y'], axis=1, inplace=True)
    Z = pd.concat([X, Y], axis=1, sort=False)
    print('Shape Z {}'.format(Z.shape))
    # SANITY check of the final Dataframe
    if (len(set([X.shape[0], Y.shape[0], Z.shape[0]] )) == 1 ):
        print('All dataframes have the same number of rows')
    else:
        raise ValueError('Dataframes are not the same size')
    print("----> Time taken to complete df_{} : {:.2f} min".format( numDays, (time.time() - t0) / 60 ) )
    
    return Z
````

Based on FIG6, the NGR will be split up as follow:
- for -ve NGR, segregate to class 0. This would be the customers we would like to remove
- for +ve NGR, cluster into 3 classes based on quantile splits.

````python
# Based on the above, the NGR will be split
print('Shape of df_ngr {}'.format(df_ngr.shape))

#  for -ve NGR, segregate to class 0
Y = df_ngr[df_ngr['GROSS_sum_ngr']<0]
Y['CLTV'] = [0]* Y.shape[0]
print('Shape of -ve NGR {}'.format(Y.shape))

# Split the remaining +ve NGRs based on 3 quantiles [1, 2, 3], with 
#     1 being the low value NGR,
#     3 being the high value NGR
X = df_ngr[df_ngr['GROSS_sum_ngr']>=0]
X['CLTV'] = pd.qcut( X['GROSS_sum_ngr'], q=3, labels=[1,2,3] )
CLTV_intervals = set(pd.qcut( X['GROSS_sum_ngr'], q=3 ))
X['CLTV'] = [ int(cltv) for cltv in X['CLTV'] ]
print('Shape of +ve NGR {}'.format(X.shape))

# concat both df
df_cltv = pd.concat([X,Y])
df_cltv.sort_values(['customer_id'], inplace=True)
print('Shape of df_cltv {}'.format(df_cltv.shape))

if (df_ngr.shape[0] == X.shape[0] + Y.shape[0]):
    print('    Shape in agreement : {} == {}'.format(df_ngr.shape[0], X.shape[0] + Y.shape[0]))
    df_ngr = df_cltv
else:
     raise ValueError('Shape not in agreement')
        
print('CLTV intervals:')
sorted(CLTV_intervals)
````
